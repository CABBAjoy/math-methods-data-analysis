### Задача 1. Коэффициент корреляции между числом единиц и числом шестёрок

**Условие:**  
Честная шестигранная игральная кость подбрасывается n раз.  
Обозначим:

- ξ — число выпаданий **единицы**,
- η — число выпаданий **шестёрки**.

Найти коэффициент корреляции между ξ и η:

$$
\rho(ξ, η) = \frac{\text{Cov}(ξ, η)}{\sqrt{D[ξ]} \cdot \sqrt{D[η]}}
$$


### Решение:

Обозначим индикаторы:

- ξ = сумма I₁ᵢ, где I₁ᵢ = 1, если на i-м броске выпала единица, иначе 0
- η = сумма I₆ᵢ, где I₆ᵢ = 1, если на i-м броске выпала шестерка

Индикаторы на одном и том же броске **зависимы** (взаимоисключающие), на разных бросках — **независимы**.

---

**Ожидания:**

- $$ E[I₁ᵢ] = E[I₆ᵢ] = \frac{1}{6} $$
- $$ E[ξ] = E[η] = n \cdot \frac{1}{6} $$

**Дисперсии:**

Для индикатора:  
$$ D[I] = p(1 - p) = \frac{1}{6} \cdot \frac{5}{6} = \frac{5}{36} $$

$$
D[ξ] = D[η] = n \cdot \frac{5}{36}
$$

---

**Ковариация:**

Поскольку ξ и η — суммы зависимых индикаторов, ковариация вычисляется:

$$
\text{Cov}(ξ, η) = \sum_{i=1}^{n} \text{Cov}(I₁ᵢ, I₆ᵢ)
$$

На **одном броске**:

$$ P(1) = P(6) = \frac{1}{6}, \quad P(1 \cap 6) = 0 $$

Тогда:

$$
\text{Cov}(I₁ᵢ, I₆ᵢ) = E[I₁ᵢ I₆ᵢ] - E[I₁ᵢ] \cdot E[I₆ᵢ] = 0 - \frac{1}{36} = -\frac{1}{36}
$$

А значит:

$$
\text{Cov}(ξ, η) = n \cdot (-\frac{1}{36}) = -\frac{n}{36}
$$

---

**Корреляция:**

$$
\rho(ξ, η) = \frac{-\frac{n}{36}}{\sqrt{n \cdot \frac{5}{36}} \cdot \sqrt{n \cdot \frac{5}{36}}} = \frac{-\frac{n}{36}}{n \cdot \frac{5}{36}} = -\frac{1}{5}
$$


### Ответ к задаче 1:

Коэффициент корреляции между числом единиц и числом шестерок:

$$
\rho(ξ, η) = \boxed{-\frac{1}{5}} = -0.2
$$


### Задача 2. Независимость случайных векторов

**Условие:**

Пусть случайные величины ξ₁, ξ₂, ξ₃, ξ₄ независимы в совокупности.  
Определим векторы:

- η = (ξ₁, ξ₂)
- ζ = (ξ₃, ξ₄)

Нужно ответить на два вопроса:

1. Независимы ли η и ζ?
2. Верно ли обратное: если η и ζ независимы, то ξ₁, ξ₂, ξ₃, ξ₄ независимы в совокупности?


### Анализ:

1. **Если ξ₁, ξ₂, ξ₃, ξ₄ независимы в совокупности**, то любые их непересекающиеся поднаборы также независимы.  
Значит: **векторы η и ζ независимы**.

2. **Обратное НЕ всегда верно**.  
Даже если векторы η и ζ независимы, это не гарантирует, что ξ₁, ξ₂, ξ₃, ξ₄ независимы **в совокупности**.  
Пример: пары могут быть независимы, но внутри каждой пары — зависимость.


### Графическая иллюстрация задачи 2

Векторы η = (X₁, X₂) и ζ = (X₃, X₄) сгенерированы из независимых источников.  
Но внутри каждой пары (X₁, X₂) и (X₃, X₄) есть сильная зависимость.

Это доказывает, что:

- Векторы η и ζ могут быть независимы,
- Однако компоненты ξ₁, ξ₂, ξ₃, ξ₄ **не обязательно независимы в совокупности**,
- Следовательно, **обратное утверждение задачи 2 — ложное**.


```python
import matplotlib.pyplot as plt
import numpy as np

# Фиксируем генератор случайных чисел
np.random.seed(42)
n = 1000

# Создаём зависимые пары внутри векторов
X1 = np.random.normal(0, 1, n)
X2 = X1 + np.random.normal(0, 0.1, n)  # Зависим от X1

X3 = np.random.normal(0, 1, n)
X4 = X3 + np.random.normal(0, 0.1, n)  # Зависим от X3

# Векторы: η = (X1, X2), ζ = (X3, X4)
# η и ζ сгенерированы независимо друг от друга

# Строим графики
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Визуализация η
axs[0].scatter(X1, X2, alpha=0.5, color='dodgerblue')
axs[0].set_title("Вектор η = (X₁, X₂) — зависимость внутри пары")
axs[0].set_xlabel("X₁")
axs[0].set_ylabel("X₂")
axs[0].grid(True)

# Визуализация ζ
axs[1].scatter(X3, X4, alpha=0.5, color='darkorange')
axs[1].set_title("Вектор ζ = (X₃, X₄) — зависимость внутри пары")
axs[1].set_xlabel("X₃")
axs[1].set_ylabel("X₄")
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
